```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os

driver = webdriver.Chrome()
driver.get("https://www.atualcard.com.br/")
time.sleep(3)

driver.find_element_by_class_name('js-form-accept-cookies').submit()
driver.find_element_by_class_name('modal__close').click()
driver.find_element_by_class_name('introjs-skipbutton').click()
driver.find_element_by_class_name('nav-user__greeting').click()
driver.find_element_by_name('login').send_keys('Seu Usuário')
driver.find_element_by_name('senha').send_keys('Sua senha')
driver.find_element_by_class_name('js-form-login').submit()
#Realizado while para que enquanto o elemento não apareça, seja aguardado 1 segundo
while len(driver.find_elements_by_xpath('/html/body/div[2]/div/main/section[3]/div/div/div/div[1]/div/a/figure/img')) == 0:
    time.sleep(1)
#Necessário após a saída do while aguardar 1 segundo, pois algumas vezes não era possível realizar a ação de clique gerando erro    
time.sleep(1)
driver.find_element_by_xpath('/html/body/div[2]/div/main/section[3]/div/div/div/div[1]/div/a/figure/img').click()
time.sleep(5)
driver.find_element_by_class_name('js-buy-product').click()
time.sleep(2)
driver.find_element_by_class_name('js-show-modal-upload').click()
while len(driver.find_elements_by_xpath('/html/body/section/div[9]/div/div/div/div[2]/form/div/label[2]/input')) == 0:
    time.sleep(1)
time.sleep(1)
driver.find_element_by_xpath('/html/body/section/div[9]/div/div/div/div[2]/form/div/label[2]/input').send_keys(os.getcwd()+r'\Koala.jpg')
time.sleep(2)
driver.find_element_by_class_name('js-upload-files').click()
time.sleep(4)
driver.find_element_by_class_name('landscape').click()
while len(driver.find_elements_by_class_name('js-hide-errors-overlay')) == 0:
    time.sleep(1)
time.sleep(1)
driver.find_element_by_class_name('js-hide-errors-overlay').click()
driver.find_element_by_class_name('js-pf-next').click()
driver.find_element_by_class_name('js-confirm-with-errors').click()
time.sleep(2)
driver.find_element_by_xpath('/html/body/section/div[8]/div/div/div[1]/div/aside/div[2]/label').click()
driver.find_element_by_class_name('js-pf-finish').click()
time.sleep(10)
#Necessário execução de comando javascript, pois ao realizar um click com comando selenium era aberta a página de termo de garantia e não era setado o checkbox
driver.execute_script('document.querySelector(".js-terms").checked = true;')
driver.find_element_by_class_name('js-to-step').click()
while len(driver.find_elements_by_xpath('/html/body/div[2]/div/main/form/div/div/div[2]/div[2]/section/div/div[1]/div/div/label[1]/span/span')) == 0:
    time.sleep(1)
time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div/main/form/div/div/div[2]/div[2]/section/div/div[1]/div/div/label[1]/span/span').click()
driver.find_element_by_xpath('/html/body/div[2]/div/main/form/div/div/div[2]/div[3]/section/div[1]/ul/li[3]/label/span').click()
driver.find_element_by_class_name('js-buy-items').click()
time.sleep(3)
driver.find_element_by_xpath('/html/body/section/div[22]/div/div/form/div/div[1]/div/input').send_keys('Usuário de Teste')
driver.find_element_by_xpath('/html/body/section/div[22]/div/div/form/div/div[2]/div/input').send_keys('41999999999')
time.sleep(2)
driver.find_element_by_xpath('/html/body/section/div[22]/div/footer/button[2]').click()
time.sleep(5)
driver.refresh()
while len(driver.find_elements_by_class_name('js-open-modal-cancel')) == 0:
    time.sleep(1)
time.sleep(1)
driver.find_element_by_class_name('js-open-modal-cancel').click()
time.sleep(1)
driver.find_element_by_name('motivoCancelamento').send_keys('teste')
driver.find_element_by_class_name('js-cancel-order').click()
time.sleep(5)
driver.quit()
```
